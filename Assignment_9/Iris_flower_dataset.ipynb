```python
#Task 1: Make a bar plot of the mean sepal sizes for each species
import matplotlib.pyplot as plt
# Load the Iris dataset
iris = seaborn.load_dataset('iris')

# Calculate the mean sepal length for each species
mean_sepal_length = iris.groupby('species')['sepal_length'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 6))
seaborn.barplot(data=mean_sepal_length, x='species', y='sepal_length')
plt.title('Mean Sepal Length for Each Species')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.show()
#seaborn.barplot(x='species', y='sepal_length', data=iris)
```

```python
# Group the data by species
#Task 2: Find the min, max and mean of petal sizes for each species
grouped = iris.groupby('species')

# Calculate the minimum, maximum, and mean of petal lengths and widths for each species
result = grouped['petal_length', 'petal_width'].agg(['min', 'max', 'mean'])

# Reset the index for a cleaner result
result = result.reset_index()

# Rename the columns for clarity
result.columns = ['Species', 'Min Petal Length', 'Max Petal Length', 'Mean Petal Length', 'Min Petal Width', 'Max Petal Width', 'Mean Petal Width']

# Display the result
print(result)
```
