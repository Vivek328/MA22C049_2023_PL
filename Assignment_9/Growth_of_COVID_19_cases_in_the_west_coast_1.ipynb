```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
```

```python
df = pd.read_csv(url)
df.head()
```

```python
if 'df' in locals() and 'Province_State' in df.columns:
    states = ['Oregon', 'Washington', 'California']
    state_data = df[df['Province_State'].isin(states)]
    num_records = len(state_data)

    print(f"Number of records in Oregon, Washington, and California: {num_records}")
else:
    print("The 'df' DataFrame or 'Province_State' column does not exist.")
```

```python
# Drop un necessary columns  in the DataFrame
columns_to_drop = ['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Admin2', 'Country_Region', 'Lat', 'Long_', 'Combined_Key']
state_data = state_data.drop(columns=[col for col in columns_to_drop if col in state_data.columns])
state_data.head()
```

```python
state_data=state_data.groupby('Province_State').sum().T
```

```python

# Plot the data for California in black with a solid line
state_data['California'].plot(color="black", linestyle='-', label='California')

# Plot the data for Washington in blue with a dashed line
state_data['Washington'].plot(color="blue", linestyle='--', label='Washington')

# Plot the data for Oregon in green with a dotted line
state_data['Oregon'].plot(color="green", linestyle=':', label='Oregon')

# Customize the labels and legend
plt.xlabel('Date')
plt.ylabel('Cases')
plt.title('COVID-19 Cases in Different States')
plt.legend()

# Show the plot
plt.show()

```
