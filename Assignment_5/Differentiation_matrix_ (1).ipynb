```python
import numpy as np
from numpy import diag
```

```python
from scipy.sparse import diags
import matplotlib.pyplot as plt
```

```python

def create_matrix(n):
    D = np.zeros((n - 1, n)) # initialization of zero numpy matrix of size (n-1) x n

    for i in range(n - 1):
        D[i, i] = -1
        D[i, i + 1] = 1

    return D

# Define a grid spacing h and x values
n = 1000
h = 2 * np.pi / (n - 1)
x_values = np.linspace(0, 2*np.pi, n)

def matrix_product_f(D, f, h):
    return np.dot(D, f) / h

# Create the matrix for the central difference method
D = create_matrix(n)

def f(x): # definition of the function f(x) = sin(x)
    return np.sin(x)

# Apply the matrix product to f(x)
df = matrix_product_f(D, f(x_values), h)

# Plot the results
plt.figure(figsize=(8, 4))
plt.title("Approximation of Derivative of f(x) = sin(x)")
plt.plot(x_values[:-1], df, label="Approximated Derivative")
plt.plot(x_values[:-1], np.cos(x_values[:-1]), label="True Derivative (cos(x))", linestyle='--')
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def f(x): # definition of the function f(x) = sin(x)
    return x**2 + 2 *np.sin(10*x)
# Actual derivative
def g(x):
  return 2*x +20 *np.cos(10*x)
# Apply the matrix product to f(x)
df = matrix_product_f(D, f(x_values), h)

# Plot the results
plt.figure(figsize=(8, 4))
plt.title("Approximation of Derivative of f(x) = x^2 + 2 sin(10x)")
plt.plot(x_values[:-1], df, label="Approximated Derivative")
plt.plot(x_values[:-1], g(x_values[:-1]), label="True Derivative 2x+20cos(10x)", linestyle='--')
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

```python
 #Define a grid spacing h and x values
n = 1000
h = 2 * np.pi / (n - 1)
x = np.linspace(0, 2*np.pi, n)
# create diff matrrix
main_diag = np.ones(n - 1)
off_diag = -np.ones(n - 1)

# Create the differentiation matrix D using scipy's diags function
D_sparse = diags([main_diag, off_diag], [1, 0], shape=(n - 1, n))
# def. of function
f_x_1 = np.sin(x)
f_x_2 = x**2 + 2 * np.sin(10*x)

# Compute the derivatives using the sparse differentiation matrix D_sparse
df1_sparse = D_sparse.dot(f_x_1) / h
df2_sparse = D_sparse.dot(f_x_2) / h
#Plot the results
plt.figure(figsize=(8, 4))
plt.title(" sparse Approximation of Derivative of f(x) = x^2 + 2 sin(10x)")
plt.plot(x[:-1],df2_sparse)
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)
plt.show()
```
