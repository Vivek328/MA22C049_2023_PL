```python
from scipy.sparse import diags
import matplotlib.pyplot as plt
```

```python
def create_matrix(n):

    D = np.zeros((n - 1, n)) # initialization of zero numpy matrix of size (n-1)xn


    for i in range(n - 1):
        D[i, i] = 1
        D[i, i + 1] = -1

    return D
A =create_matrix(4)
print(A)

def matrix_product_f(D, f):
    return np.dot(D, f)

n = 5 #Create  matrix when  n=5
D = create_matrix(n)


def f(x): # definition of function of f(x)=sin(x)
    return np.sin(x)

# Define a grid spacing h and x values
h = 0.1
x_values = np.arange(0, n * h, h)

# Apply the  matrix product to f(x)
df = matrix_product_f(D, f(x_values))

# Plot the results
plt.figure(figsize=(8, 4))
plt.title("Approximation of Derivative of f(x) = sin(x)")
plt.plot(x_values[:-1], df, label="Approximated Derivative")
plt.plot(x_values[:-1], np.cos(x_values[:-1]), label="True Derivative (cos(x))", linestyle='--')
plt.xlabel("x")
plt.ylabel("Derivative")
plt.legend()
plt.grid(True)
plt.show()
```
