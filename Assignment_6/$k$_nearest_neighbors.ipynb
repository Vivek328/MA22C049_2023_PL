```python
import numpy as np

#  define the Function to calculate Euclidean distance between two points
def euclidean_distance(point1, point2):
    return np.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)
```

```python
# Function to find k nearest neighbors for each point
def k_nearest(points, k):
 # points: function  taking list of  points for which you want to find the nearest neighbours.
 #  k: This is an integer representing the number of nearest neighbors to find for each point.
    k_nearest_neighbours = [] # creating empty list to store k_nearest points for each point
    for point in points:
      distances=[]
      for other_point in points:
        distances.append((other_point, euclidean_distance(point, other_point)))

      distances.sort(key=lambda x: x[1])  # Sort by distance
      k_nearest = [neighbour[0] for neighbour in distances[1:k + 1]]  #  here neighbour[0] is used to extract first element of distance sort and Exclude the point itself
      k_nearest_neighbours.append(k_nearest)
    return k_nearest_neighbours
```

```python
# Given set of points and k
P = [[0, 0], [0.2, 0.22], [0.1, -0.1], [1, 1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63],
     [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8, -0.23], [0.8, 0.6]]
k = 3

# Call the function to find k nearest neighbors
k_nearest_neighbours = k_nearest(P, k)
```

```python
# Function to plot arrows with direction
def plot_arrows(points, neighbours):
    for i, point in enumerate(points):
        for neighbour in neighbours[i]:
            plt.annotate("", xy=neighbour, xytext=point,arrowprops={})
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))

# Plot the points
x, y = zip(*P)
plt.scatter(x, y, c='red', marker='o', label='Points')

# Plot arrows with direction from each point to its k-nearest neighbors
plot_arrows(P, k_nearest_neighbours)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('K-Nearest Neighbours for K=3 ')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
