```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
def f(x):
  return np.sin(x)
x_values=np.linspace(0,10,num=1000)
y_values =np.sin(x_values[:-1])
# plot of Graph sin(x)
plt.figure(figsize=(10,8))
plt.title("Graph  of f(x) = sin(x)")
plt.plot(x_values[:-1], y_values)
plt.plot(x_values[:-1], y_values)
plt.xlabel("x")
plt.ylabel("y_values of sin(x)")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm  # colour map
```

```python
def g(x,y):
  return np.cos(np.sqrt(x**2+ y**2))
x=np.linspace(-5,5,num=1000)
y=np.linspace(-5,5,num=1000)
X,Y =np.meshgrid(x , y)
Z=g(X,Y)
fig = plt.figure() # create a new figure
ax = plt.axes( projection='3d')
ax.set_title("Surface of z=cos((x^2 + y^2)^1/2)")
ax.set_xlabel("x values ")
ax.set_ylabel("y values")
ax.set_zlabel("z values ")

ax.grid(True)

surf = ax.plot_surface(X, Y, Z, cmap=cm.gist_rainbow )
plt.colorbar(surf)
#plt.savefig('Graphing fun.png')
plt.show()

```
