```python
#eigenvalue decomposition of an n × n matrix A is of the form
#A = XDX^−1
#where D is a diagonal matrix and X is an invertible matrix, both of the same size as A. The
#eigenvalues of A are found in the diagonal entries of D and the eigenvectors are columns of X
# svd :SVD is a mathematical technique used for decomposing a matrix into three other matrices
```

```python
from scipy.linalg import svd  #  import svd function from the scipy.linalg  module
import numpy as np
# This  line of code that customizes how NumPy (the numpy library) will display numbers
np.set_printoptions(precision=3, suppress=True)
#precision=3: This argument specifies that when you print floating-point numbers, they should be displayed with a precision of 3 decimal places
#suppress=True: This argument tells NumPy to suppress small numbers in scientific notation
```

```python
# creates a random 4x5 matrix a with complex numbers
a = np.random.rand(4, 5) + 1j * np.random.rand(4, 5)
#performs the SVD on the matrix 'a' and stores the results in three variables: u, s, and vh.
u, s, vh = svd(a)
```

```python
# Display the results
print("Matrix a:")
print(a)
print("\nLeft Singular Vector Matrix u:")
print(u)
print("\nSingular Values (s):")
print(s)
print("\nConjugate Transpose of Right Singular Vector Matrix (vh):")
print(vh)
```

```python
u @ u.T.conjugate() # u is unitary. Its columns are left singular vectors
```

```python
vh @ vh.T.conjugate() # Rows of vh are right singular vectors
```

```python
s # Only the diagonal entries of Sigma are returned in s
```

```python
a = np.random.rand(4, 5)
u, s, vh = svd(a)
```

```python
u[0, :, np.newaxis] @ vh[np.newaxis, 0, :] # matrix multiplication ,  u[0, :]does the job  selects the first row of matrix u
#np.newaxis is used to add an extra dimension to the selected row, effectively converting it from a 1D array to a 2D column vector
```

```python
np.outer(u[0, :], vh[0, :])
```

```python
ar = np.zeros_like(a) #     initilization of  zero matrix like matrix a
for i in range(4): #u[:, i] represents the i th column of the 'u' matrix
     ar += np.outer(u[:, i], s[i] * vh[i, :])
```

```python
a - ar # a and ar are identical
```

```python
relative_error = 1.e-1
```

```python
s2 = s**2
total = np.sum(s2)
diff = np.sqrt((total - np.add.accumulate(s2)) / total)
l = np.argmax(diff < relative_error) + 1
l
```

```python
cl = u[:, :l] @ np.diag(s[:l]) @ vh[:l, :]
```

```python
np.linalg.norm(c - cl, 'fro') / np.linalg.norm(c, 'fro')
```
